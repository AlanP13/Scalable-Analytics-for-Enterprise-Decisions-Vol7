```python
# Import the libraries we will use in this lab.

# pandas is used for data loading, manipulation, and CSV handling
import pandas as pd   
# json is used to convert between JSON strings and Python objects
import json           

# The following imports are for the NBA API part (Part 3).
from nba_api.stats.static import players, teams         # used to look up player and team IDs
from nba_api.stats.endpoints import PlayerGameLog       # used to retrieve game logs for a player
```

```python
# Part 1: CSV Data Handling
print("="*100)
print("Part 1: CSV Data Handling")
print("="*100)

# Step 1 & 2: Load the Kaggle dataset 'winequality-red.csv' using pandas.
# IMPORTANT: Make sure the file 'winequality-red.csv' is in the same folder as this notebook.
# If it is somewhere else, update the path below accordingly.

csv_path = r'C:\Users\alanp\Downloads\winequality-red.csv'  # relative path to the CSV file

# Read the CSV into a pandas DataFrame.
df_wine = pd.read_csv(csv_path)

# Print a message and basic information to confirm that the data loaded correctly.
print("Loaded winequality-red.csv successfully.")
print("Shape of original DataFrame (rows, columns):", df_wine.shape)

# Display the first 5 rows to see what the data looks like.
df_wine.head()
```

```python
# Step 3: Add a new column to assign a unique ID to each row.
# We want an 'id' column that starts at 1 and increments by 1 for each row.

# range(1, len(df_wine) + 1) generates a sequence: 1, 2, 3, ..., number_of_rows.
# df_wine.insert(0, "id", ...) inserts this as the first column (position 0).
df_wine.insert(0, "id", range(1, len(df_wine) + 1))

# Step 4: Display the first 10 rows of the updated DataFrame
# to verify that the 'id' column was added correctly.
print("First 10 rows of the updated DataFrame (with 'id' column added):")
df_wine.head(10)
```

```python
# Step 5: Export the modified DataFrame to a new CSV file using to_csv.

# Choose a name for the new CSV file that clearly shows it includes the ID column.
output_csv_path = r'C:\Users\alanp\Downloads\winequality-red_with_id.csv'

# Save the DataFrame to CSV.
# index=False means we do NOT want to write the DataFrame index as a separate column in the file.
df_wine.to_csv(output_csv_path, index=False)

# Print a confirmation message.
print(f"Modified DataFrame saved to: {output_csv_path}")
```

```python
# Part 2: JSON Operations
print("="*100)
print("Part 2: JSON Operations")
print("="*100)
# Step 7: Convert a sample JSON string into a Python object.

# Define a JSON-formatted string. This is how data would look if it came from an API or file.
sample_json_str = '{"name": "Alan", "age": 25, "is_student": true, "courses": ["Data Science", "Cybersecurity", "AI"]}'

# json.loads() takes a JSON string and converts it to a Python object.
# In this case, it will return a Python dictionary.
python_obj_from_json = json.loads(sample_json_str)

print("7) JSON string -> Python object")
print("Original JSON string:", sample_json_str)
print("Converted Python object:", python_obj_from_json)
print("Python object type:", type(python_obj_from_json))  # should be <class 'dict'>
```

```python
# Step 8: Convert a Python object into JSON data (a JSON string).

# Create a regular Python dictionary.
python_obj = {
    "course": "Big Data Analytics",
    "credits": 3,
    "online": True
}

# json.dumps() converts Python objects (dicts, lists, etc.) into JSON-formatted strings.
json_from_python = json.dumps(python_obj)

print("8) Python object -> JSON string")
print("Original Python object:", python_obj)
print("Converted JSON string:", json_from_python)
print("JSON string type:", type(json_from_python))  # should be <class 'str'>
```

```python
# Step 9: Convert various Python objects into JSON strings and print the outputs.

# Define different types of Python objects.
var_int = 42
var_list = [1, 2, 3, "four"]
var_tuple = ("a", "b", "c")          # JSON does not have tuples, so this becomes a JSON array
var_dict = {"x": 10, "y": 20}
var_bool = False

print("9) Various Python objects -> JSON strings:\n")

# Use json.dumps() on each object type.
print("Integer:", var_int, "-> JSON:", json.dumps(var_int))
print("List:", var_list, "-> JSON:", json.dumps(var_list))
print("Tuple:", var_tuple, "-> JSON:", json.dumps(var_tuple))   # converted to list in JSON
print("Dictionary:", var_dict, "-> JSON:", json.dumps(var_dict))
print("Boolean:", var_bool, "-> JSON:", json.dumps(var_bool))
```

```python
# Step 10: Convert a Python dictionary (sorted by keys) into a formatted JSON string.

# Define a dictionary with keys out of order.
dict_for_pretty_json = {
    "beta": 2,
    "alpha": 1,
    "delta": 4,
    "gamma": 3
}

# json.dumps with sort_keys=True sorts the dictionary keys alphabetically.
# indent=4 formats the JSON string across multiple lines with indentation,
# making it easier to read.
pretty_json = json.dumps(dict_for_pretty_json, sort_keys=True, indent=4)

print("10) Sorted & pretty-printed JSON (indent=4):")
print(pretty_json)
```

```python
# Part 3: Working with NBA API
print("="*100)
print("Part 3: Working with NBA API")
print("="*100)

print("Using nba_api to look up player and team information...")

# Step 12: Get a list of all NBA players from the nba_api.
# players.get_players() returns a list of dictionaries,
# where each dictionary represents a player.
player_dict = players.get_players()

# We want to find the dictionary for LeBron James.
# This list comprehension filters the list to players whose full_name matches exactly.
bron = [p for p in player_dict if p["full_name"] == "LeBron James"][0]

# Extract LeBron's unique player ID from the dictionary.
bron_id = bron["id"]

print("Player info for:")
print(bron)          # print the full player dictionary
print("Player ID:", bron_id)

# Now we do something similar for teams.
teams_list = teams.get_teams()  # returns a list of team dictionaries

# Filter the list to find the Golden State Warriors.
GSW = [t for t in teams_list if t["full_name"] == "Golden State Warriors"][0]

# Extract the team ID for GSW.
GSW_id = GSW["id"]

print("\nTeam info for Golden State Warriors:")
print(GSW)
print("GSW ID:", GSW_id)
```

```python
# Step 13: Collect game data for a player of your choice and export it to CSV.
# In this example, we will use Stephen Curry (using bron_id from above).
player_dict = players.get_players()

bron = [p for p in player_dict if p["full_name"] == "Stephen Curry"][0]

# Extract LeBron's unique player ID from the dictionary.
bron_id = bron["id"]

print("Player info for:")
print(bron)          # print the full player dictionary
print("Player ID:", bron_id)
print("Fetching game log data for Stephen Curry (Regular Season 2023-24)...")

# PlayerGameLog is an endpoint in nba_api that retrieves game-by-game stats for a player.
# season="2023-24" specifies the NBA season.
# season_type_all_star="Regular Season" filters to regular season games only.
gamelog = PlayerGameLog(
    player_id=bron_id,
    season="2023-24",
    season_type_all_star="Regular Season"
)

# get_data_frames() returns a list of pandas DataFrames.
# The first element [0] contains the main game log.
df_games = gamelog.get_data_frames()[0]

print("First 5 rows of Stephen Curry's game log data:")
df_games.head()
```

```python
# Export the game log DataFrame to a CSV file so it can be submitted.

# Choose a filename that reflects the player and season.
example_nba_csv_path = r'C:\Users\alanp\Downloads\lebron_games_2023_24.csv'
nba_csv_path = r'C:\Users\alanp\Downloads\curry_games_2023_24.csv'

# Save the DataFrame to CSV without the index.
df_games.to_csv(example_nba_csv_path, index=False)
df_games.to_csv(nba_csv_path, index=False)

print(f"Game data saved to: {example_nba_csv_path}")
print(f"Game data saved to: {nba_csv_path}")
```
